7-2 两个有序链表序列的交集

已知两个非降序链表序列S1与S2，设计函数构造出S1与S2的交集新链表S3。

输入格式:
输入分两行，分别在每行给出由若干个正整数构成的非降序序列，用−1表示序列的结尾（−1不属于这个序列）。数字用空格间隔。

输出格式:
在一行中输出两个输入序列的交集序列，数字间用空格分开，结尾不能有多余空格；若新链表为空，输出NULL。

输入样例:
```
1 2 5 -1
2 4 5 8 10 -1
```
输出样例:
```
2 5
```

````python
import functools as f

def getInput():
    return input().split()

def StrToInt(arr :list):
    return [ int(i) for i in arr]

In = f.partial(lambda: StrToInt(getInput()))

if __name__ == "__main__":

    UserInput = {i: In() for i in range(2)}

    sets = [set(v) for v in UserInput.values() if v is not None]
    
    if sets:
        inter2 = set.intersection(*sets)
    else:
        inter2 = set()

    inter2.remove(-1)

    if len(inter2) == 0:
        print("NULL",end="")
    else:
        lst = list(inter2)
        print(" ".join(map(str, lst)))
````

````python
# ...existing code...
import sys
from typing import Iterator, Optional

def gen_line() -> Iterator[int]:
    """
    从 stdin 读取一行，按空格分解为整数，遇到 -1 停止（-1 不属于序列）。
    注意：此实现对单行仍调用 split，会生成临时列表；若行极大需改为字符流解析。
    """
    line = sys.stdin.readline()
    if not line:
        return
    for x in map(int, line.split()):
        if x == -1:
            break
        yield x

def intersection_stream(a: Iterator[int], b: Iterator[int]) -> Iterator[int]:
    """
    在两个非降序整数流上以双指针方式求交集，去重并按非降序输出。
    产出顺序稳定，额外内存 O(1)。
    """
    try:
        ai = next(a)
    except StopIteration:
        ai = None
    try:
        bi = next(b)
    except StopIteration:
        bi = None

    last: Optional[int] = None
    while ai is not None and bi is not None:
        if ai == bi:
            if last != ai:
                yield ai
                last = ai
            # 两边都前进
            try:
                ai = next(a)
            except StopIteration:
                ai = None
            try:
                bi = next(b)
            except StopIteration:
                bi = None
        elif ai < bi:
            try:
                ai = next(a)
            except StopIteration:
                ai = None
        else:
            try:
                bi = next(b)
            except StopIteration:
                bi = None

def main():
    a = gen_line()
    b = gen_line()

    first_printed = False
    for val in intersection_stream(a, b):
        if not first_printed:
            # 第一个元素直接打印（不前置空格）
            print(val, end='')
            first_printed = True
        else:
            # 其余元素前加空格
            print(' ' + str(val), end='')

    if not first_printed:
        # 没有输出任何元素，按题目要求输出 NULL（末尾换行）
        print("NULL",end="")
    else:
        # 已输出元素，补一个换行
        print()

if __name__ == "__main__":
    main()
# ...existing code...
````

````python
#include <bits/stdc++.h>
using namespace std;

// 从一行读取整数序列，遇到 -1 停止（-1 不属于序列）
vector<int> read_sequence_line() {
    string line;
    if (!getline(cin, line)) return {};
    istringstream iss(line);
    int x;
    vector<int> v;
    while (iss >> x) {
        if (x == -1) break;
        v.push_back(x);
    }
    return v;
}

// 在两个非降序向量上用双指针求交集（去重），直接输出，避免多余存储
void output_intersection(const vector<int>& a, const vector<int>& b) {
    size_t i = 0, j = 0;
    bool printed = false;
    int last = numeric_limits<int>::min(); // 记录上次已输出值以去重

    while (i < a.size() && j < b.size()) {
        if (a[i] == b[j]) {
            if (!printed || a[i] != last) {
                if (printed) cout << ' ';
                cout << a[i];
                printed = true;
                last = a[i];
            }
            ++i; ++j;
        } else if (a[i] < b[j]) {
            ++i;
        } else {
            ++j;
        }
    }

    if (!printed) cout << "NULL";
    if (printed) cout << '\n';
}

int main() {
    ios::sync_with_stdio(false);
    cin.tie(nullptr);

    // 读取两行序列
    vector<int> a = read_sequence_line();
    vector<int> b = read_sequence_line();

    output_intersection(a, b);
    return 0;
}
````

````python
#include <bits/stdc++.h>
using namespace std;

struct ListNode {
    int val;
    ListNode* next;
    ListNode(int x): val(x), next(nullptr) {}
};

ListNode* buildList(const vector<int>& v) {
    if (v.empty()) return nullptr;
    ListNode* head = new ListNode(v[0]);
    ListNode* cur = head;
    for (size_t i = 1; i < v.size(); ++i) {
        cur->next = new ListNode(v[i]);
        cur = cur->next;
    }
    return head;
}

// 构造两个非降序链表的交集链表（每个值在结果中只出现一次）
ListNode* intersectLists(ListNode* a, ListNode* b) {
    ListNode dummy(0);
    ListNode* tail = &dummy;
    while (a && b) {
        if (a->val < b->val) a = a->next;
        else if (a->val > b->val) b = b->next;
        else {
            if (tail == &dummy || tail->val != a->val) {
                tail->next = new ListNode(a->val);
                tail = tail->next;
            }
            int v = a->val;
            while (a && a->val == v) a = a->next;
            while (b && b->val == v) b = b->next;
        }
    }
    return dummy.next;
}

int main() {
    string line;
    vector<int> v1, v2;
    if (!getline(cin, line)) return 0;
    {
        istringstream ss(line);
        int x;
        while (ss >> x) {
            if (x == -1) break;
            v1.push_back(x);
        }
    }
    if (!getline(cin, line)) return 0;
    {
        istringstream ss(line);
        int x;
        while (ss >> x) {
            if (x == -1) break;
            v2.push_back(x);
        }
    }

    ListNode* s1 = buildList(v1);
    ListNode* s2 = buildList(v2);
    ListNode* s3 = intersectLists(s1, s2);

    if (!s3) {
        cout << "NULL";
    } else {
        bool first = true;
        for (ListNode* p = s3; p; p = p->next) {
            if (!first) cout << ' ';
            cout << p->val;
            first = false;
        }
        cout << '\n';
    }
    return 0;
}
````
