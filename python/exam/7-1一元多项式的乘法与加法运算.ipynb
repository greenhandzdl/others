数字间以空格分隔。

输出格式:
输出分2行，分别以指数递降方式输出乘积多项式以及和多项式非零项的系数和指数。
数字间以空格分隔，但结尾不能有多余空格。零多项式应输出0 0。

输入样例:
```bash
4 3 4 -5 2  6 1  -2 0
3 5 20  -7 4  3 1
```
输出样例:
```bash

5 20 -4 4 -5 2 9 1 -2 0
```

````python
import functools as f
import itertools as Iter

def getInput():
    return input().split()

def StrLToIntL(arr: list):
    return [int(i) for i in arr]

In = f.partial(lambda: StrLToIntL(getInput()))

def HandlePol(arr: list):
    length = len(arr)
    # 修改1: 跳过第一个元素（项数），从第二个元素开始处理
    return {arr[i + 1]: arr[i] for i in range(1, length - 1, 2)}

def AddPol(Pol1: dict, Pol2: dict):
    coeff = set(Pol1.keys()) | set(Pol2.keys())
    return {exp: Pol1.get(exp, 0) + Pol2.get(exp, 0) 
            for exp in coeff 
            if Pol1.get(exp, 0) + Pol2.get(exp, 0) != 0}

def MulPol(Pol1: dict, Pol2: dict):
    temp = {}   
    for (exp1, coeff1), (exp2, coeff2) in Iter.product(Pol1.items(), Pol2.items()):
        new_exp = exp1 + exp2
        new_coeff = coeff1 * coeff2
        if new_exp in temp:
            temp[new_exp] += new_coeff
        else:
            temp[new_exp] = new_coeff

    # 修改2: 按指数降序排列，而不是升序
    sorted_items = sorted(temp.items(), key=lambda x: x[0], reverse=True)
    return {exp: coeff for exp, coeff in sorted_items if coeff != 0}

def OpPol(Pol1: dict, Pol2: dict):
    return AddPol(Pol1, Pol2), MulPol(Pol1, Pol2)

def ShowPol(Pol: dict):
    # 修改3: 处理零多项式情况
    if not Pol:
        print("0 0")
        return
    
    # 修改4: 正确格式化输出，避免尾随空格
    sorted_items = sorted(Pol.items(), key=lambda x: x[0], reverse=True)
    output = []
    for exp, coeff in sorted_items:
        output.append(f"{coeff} {exp}")
    print(" ".join(output))

if __name__ == "__main__":
    ListPara = {i: In() for i in range(2)}
    
    # 测试数据
    # ListPara = {
    #     0: StrLToIntL("4 3 4 -5 2  6 1  -2 0".split()),
    #     1: StrLToIntL("3 5 20  -7 4  3 1".split())
    # }
    
    HandledPol = {i: HandlePol(ListPara[i]) for i in ListPara}
    
    res = OpPol(HandledPol[0], HandledPol[1])
    
    # 修改5: 分别输出乘积和和，而不是用reversed
    ShowPol(res[1])  # 乘积
    ShowPol(res[0])  # 和
````
