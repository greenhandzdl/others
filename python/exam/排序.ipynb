本题要求从键盘读入n个整数，对这些数做选择排序。输出选择排序每一步的结果和最终结果。

输入格式:
输入的第一行是一个正整数n，表示 在第二行中会有n个整数。

输出格式:
输出选择排序每一步的结果和最终结果。

输入样例:
在这里给出一组输入。例如：
```
5
3 7 2 9 1
```
输出样例:
在这里给出相应的输出。例如：
```
step 1: 1 7 2 9 3 
step 2: 1 2 7 9 3 
step 3: 1 2 3 9 7 
step 4: 1 2 3 7 9 
sorted array: 1 2 3 7 9 
```
注意：
输出的冒号 : 是英文输入法下的符号，冒号后有一个空格。每个整数后有一个空格。

````python
import sys as sys

Input = sys.stdin.read().splitlines()
InputIter = iter(Input)


def SelectionSort(arr):
    '''
    Selection Sort

    Parameters:
    arr: list of integers

    Returns:
    array and swap it with the element at the current position.
    '''
    count = 0
    for i in range(len(arr)):
        min_index = i
        for j in range(i + 1, len(arr)):
            if arr[j] < arr[min_index]:
                min_index = j
        if count > 0:
            print(f"step {count}: {' '.join(map(str, arr))} ")
        count += 1
        arr[i], arr[min_index] = arr[min_index], arr[i]

if __name__ == '__main__':
    _ = next(InputIter) # Skip the first line
    
    arr =  [ int(i) for i in next(InputIter).split() ] # Read the array of integers

    SelectionSort(arr)
    print(f"sorted array: {' '.join(map(str, arr))}", end=" ")
````

根据维基百科的定义：

插入排序是迭代算法，逐一获得输入数据，逐步产生有序的输出序列。每步迭代中，算法从输入序列中取出一元素，将之插入有序序列中正确的位置。如此迭代直到全部元素有序。

堆排序也是将输入分为有序和无序两部分，迭代地从无序部分找出最大元素放入有序部分。它利用了大根堆的堆顶元素最大这一特征，使得在当前无序区中选取最大元素变得简单。

现给定原始序列和由某排序算法产生的中间序列，请你判断该算法究竟是哪种排序算法？

输入格式：
输入在第一行给出正整数 n (≤100)；随后一行给出原始序列的 n 个整数；最后一行给出由某排序算法产生的中间序列。这里假设排序的目标序列是升序。数字间以空格分隔。

输出格式：
首先在第 1 行中输出 Insertion Sort 表示插入排序、或 Heap Sort 表示堆排序；然后在第 2 行中输出用该排序算法再迭代一轮的结果序列。题目保证每组测试的结果是唯一的。数字间以空格分隔，且行首尾不得有多余空格。

输入样例 1：
```
10
3 1 2 8 7 5 9 4 6 0
1 2 3 7 8 5 9 4 6 0
```
输出样例 1：
Insertion Sort
1 2 3 5 7 8 9 4 6 0
输入样例 2：
```
10
3 1 2 8 7 5 9 4 6 0
6 4 5 1 0 3 2 7 8 9
```
输出样例 2：
Heap Sort
5 4 3 1 0 2 6 7 8 9

````python
#include <iostream>
#include <vector>
#include <algorithm>
using namespace std;

// 堆调整函数
void heapAdjust(vector<int>& arr, int start, int end) {
    int parent = start;
    int child = 2 * parent + 1;  // 左孩子
    
    while (child <= end) {
        // 如果有右孩子且右孩子比左孩子大
        if (child + 1 <= end && arr[child] < arr[child + 1]) {
            child++;
        }
        
        // 如果父节点大于等于孩子节点，调整结束
        if (arr[parent] >= arr[child]) {
            return;
        }
        
        // 交换父节点和孩子节点
        swap(arr[parent], arr[child]);
        parent = child;
        child = 2 * parent + 1;
    }
}

int main() {
    int n;
    cin >> n;
    
    vector<int> original(n), current(n);
    
    // 读取原始序列
    for (int i = 0; i < n; i++) {
        cin >> original[i];
    }
    
    // 读取中间序列
    for (int i = 0; i < n; i++) {
        cin >> current[i];
    }
    
    // 判断排序类型
    // 找到第一个不满足递增的位置
    int i = 1;
    while (i < n && current[i] >= current[i - 1]) {
        i++;
    }
    
    // 检查从i开始的部分是否与原始序列相同
    bool isInsertion = true;
    int start = i;
    while (i < n) {
        if (current[i] != original[i]) {
            isInsertion = false;
            break;
        }
        i++;
    }
    
    if (isInsertion) {
        cout << "Insertion Sort" << endl;
        
        // 插入排序的下一轮：将start位置的元素插入到前面已排序部分
        vector<int> next = current;
        int temp = next[start];
        int j = start - 1;
        while (j >= 0 && next[j] > temp) {
            next[j + 1] = next[j];
            j--;
        }
        next[j + 1] = temp;
        
        // 输出结果
        for (int k = 0; k < n; k++) {
            if (k > 0) cout << " ";
            cout << next[k];
        }
        // cout << endl;
    } else {
        cout << "Heap Sort" << endl;
        
        // 堆排序的下一轮
        // 从后往前找到第一个不满足递减的位置
        i = n - 1;
        while (i > 0 && current[i] >= current[i - 1]) {
            i--;
        }
        
        // 此时i指向堆的最后一个元素
        vector<int> next = current;
        
        // 交换堆顶元素和堆的最后一个元素
        swap(next[0], next[i]);
        
        // 调整堆
        heapAdjust(next, 0, i - 1);
        
        // 输出结果
        for (int k = 0; k < n; k++) {
            if (k > 0) cout << " ";
            cout << next[k];
        }
        // cout << endl;
    }
    
    return 0;
}
````
