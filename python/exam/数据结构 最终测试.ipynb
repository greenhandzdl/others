### 测试1:抢红包

没有人没抢过红包吧•这里给出N个人之间互相发红包、抢红包的记录，请你统计一下他们抢红包的收获。

输入格式：

输入第一行给出一个正整数N（≤104），即参与发红包和抢红包的总人数，则这些人从1到N编号。随后N行，第i行给出编号为i的人发红包的记录，格式如下：

K NI PI ... NK PK

其中K（0≤K ≤20）是发出去的红包个数，N：是抢到红包的人的编号，P（>0）是其抢到的红包金额（以分为单位）。注意：对于同一个人发出的红包，每人最多只能抢1次， 不能重复抢。

输出格式：

按照收入金额从高到低的递减顺序输出每个人的编号和收入金额（以元为单位，输出小数点后2位）。每个人的信息占一行， 两数字间有1个空格。如果收入金额有并列，则按抢到红包的个数递减输出；如果还有并列，则按个人编号递增输出。

```python
import sys as sys
lines = sys.stdin.read().splitlines()
if not lines:
    print("No input provided")
    sys.exit(0)
try:
    n = int(lines[0].strip())
except Exception:
    print("Invalid input")
    sys.exit(0)
money = [0] * (n + 1)  # cents
received_count = [0] * (n + 1)
# Process next n lines, each line i corresponds to giver i (1-indexed)
for i in range(1, n + 1):
    if i >= len(lines):
        break
    line = lines[i].strip()
    if not line:
        continue
    parts = line.split()
    # First element is K
    try:
        k = int(parts[0])
    except Exception:
        continue
    # Expect 2*k numbers following
    for j in range(k):
        idx = 1 + 2 * j
        if idx + 1 >= len(parts):
            break
        try:
            receiver = int(parts[idx])
            amount = int(parts[idx + 1])
        except Exception:
            continue
        if 1 <= receiver <= n:
            money[receiver] += amount
            money[i] -= amount
            received_count[receiver] += 1
# Prepare IDs and sort: money desc, received_count desc, id asc
ids = list(range(1, n + 1))
ids.sort(key=lambda x: (-money[x], -received_count[x], x))
for idx in ids:
    # print in yuan with two decimals
    print(f"{idx} {money[idx]/100:.2f}")
```

给定一棵二叉树的后序遍历和中序遍历，请你输出其层序遍历的序列。这里假设键值都是互不相等的正整数。

输入格式：

输入第一行给出一个正整数N（≤30），是二叉树中结点的个数。第二行给出其后序遍历序列。第三行给出其中序遍历序列。

数字间以空格分隔。

输出格式：

在一行中输出该树的层序遍历的序列。数字间以1个空格分隔， 行首尾不得有多余空格。

```python
import sys as sys
# 构建树
def buildTree(PosRead, InRead):
	if not PosRead or not InRead:
		return None

	# 后序遍历的最后一个节点是根节点
	rootValue = PosRead[-1]
	root = Tree(rootValue)

	# 在中序遍历中找到根节点的位置
	rootIndex = InRead.index(rootValue)

	# 构建左子树和右子树
	root.left = buildTree(PosRead[:rootIndex], InRead[:rootIndex])
	root.right = buildTree(PosRead[rootIndex:-1], InRead[rootIndex + 1:])

	return root
class Tree:
	def __init__(self, value):
		self.value = value
		self.left = None
		self.right = None

	def traverse_as_hierarchical(self):
		'''层序遍历'''''
		if not self:
			return ''
		result = []
		queue = [self]
		while queue:
			current = queue.pop(0)
			result.append(str(current.value))
			if current.left:
				queue.append(current.left)
			if current.right:
				queue.append(current.right)
		return ' '.join(result)
if __name__ == '__main__':
	# Read Nodes
	NodeNum = int(sys.stdin.readline().strip())

	# 后序遍历
	PosRead = list(map(int, sys.stdin.readline().strip().split()))

	# 中序遍历
	InRead = list(map(int, sys.stdin.readline().strip().split()))

	root = buildTree(PosRead, InRead)

	print(root.traverse_as_hierarchical())
```

图着色问题是一个著名的NP完全问题。给定无向图G=(V,E)，问可否用K种颜色为V中的每一个顶点分配一种颜色，使得不会有两个相邻顶点具有同一种颜色？

但本题并不是要你解决这个着色问题，而是对给定的一种颜色分配，请你判断这是否是图着色问题的一个解。

输入格式：
输入在第一行给出3个整数V（0<V≤500）、E（≥0）和K（0<K≤V），分别是无向图的顶点数、边数、以及颜色数。顶点和颜色都从1到V编号。随后E行，每行给出一条边的两个端点的编号。在图的信息给出之后，给出了一个正整数N（≤20），是待检查的颜色分配方案的个数。随后N行，每行顺次给出V个顶点的颜色（第i个数字表示第i个顶点的颜色），数字间以空格分隔。题目保证给定的无向图是合法的（即不存在自回路和重边）。

输出格式：
对每种颜色分配方案，如果是图着色问题的一个解则输出Yes，否则输出No，每句占一行。

```python
import sys
def solve():
    # Use fast I/O by reading all input at once
    input_data = sys.stdin.read().split()
    if not input_data:
        return

    ptr = 0
    V = int(input_data[ptr])
    E = int(input_data[ptr+1])
    K = int(input_data[ptr+2])
    ptr += 3

    # Store edges as a list of tuples
    # Since we only need to check each edge once per query,
    # a simple list of edges is actually O(E) per query.
    # The bottleneck in the previous version might have been slow I/O (readline).
    edges = []
    for _ in range(E):
        u = int(input_data[ptr])
        v = int(input_data[ptr+1])
        edges.append((u - 1, v - 1))
        ptr += 2

    N = int(input_data[ptr])
    ptr += 1

    results = []
    for _ in range(N):
        # Read V colors for this query
        colors = input_data[ptr : ptr + V]
        ptr += V

        # Check 1: Number of distinct colors must be exactly K
        # Using a set of strings is faster than converting to int first
        distinct_colors = set(colors)
        if len(distinct_colors) != K:
            results.append("No")
            continue

        # Check 2: No two adjacent vertices have the same color
        is_valid = True
        for u, v in edges:
            if colors[u] == colors[v]:
                is_valid = False
                break

        if is_valid:
            results.append("Yes")
        else:
            results.append("No")

    # Batch output for better performance
    sys.stdout.write("\n".join(results))
if __name__ == "__main__":
    solve()
```

输入格式：
输入在第一行给出两个正整数 N 和 M（均不超过10 000），分别为敌方城市个数（于是默认城市从 1 到 N 编号）和连接两城市的通路条数。随后 M 行，每行给出一条通路所连接的两个城市的编号，其间以一个空格分隔。在城市信息之后给出参谋部的系列方案，即一个正整数 K （≤ 100）和随后的 K 行方案，每行按以下格式给出：

Np v[1] v[2] ... v[Np]
其中 Np 是该方案中计划攻下的城市数量，后面的系列 v[i] 是计划攻下的城市编号。

输出格式：
